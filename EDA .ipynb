```python
import pandas as pd

# Load CSV file
df = pd.read_csv("Clean_Financial_Sample.csv")

# Currency columns to clean
currency_cols = ['Discounts', 'Sales', 'COGS', 'Profit']

# Safe cleaning function
def clean_currency_column(col):
    col = col.astype(str)  # ensure string type
    col = col.replace(r'\((.*?)\)', r'-\1', regex=True)  # handle negative in brackets
    col = col.replace(r'[-â€“\s]*$', '0', regex=True)  # treat dashes and spaces as 0
    col = col.replace('[\$,]', '', regex=True)  # remove $ and commas
    col = col.replace('^\s*$', '0', regex=True)  # replace blanks with 0
    return col.astype(float)

# Apply function to each currency column
for col in currency_cols:
    df[col] = clean_currency_column(df[col])

# Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')  # 'coerce' handles invalid dates

# Optional: Save cleaned file
df.to_csv("Cleaned_Financial_Sample.csv", index=False)

# Preview cleaned data
print(df.head())
```
